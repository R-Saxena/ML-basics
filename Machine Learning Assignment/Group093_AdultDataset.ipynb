### Treatment of outliers

```python
def outlier_treatment(df, column):
    upper_limit = df[column].mean() + 3*df[column].std()
    lower_limit = df[column].mean() - 3*df[column].std()
    
    print("Total outliers detected in "+ column + " are " + str(df[(df[column] > upper_limit) | (df[column] < lower_limit)].shape[0]))
    df[column] = np.where(
        df[column]>upper_limit,
        upper_limit,
        np.where(
            df[column]<lower_limit,
            lower_limit,
            df[column]
        )
    )
    
    return
    
```

###### For the outliers handling, we are capping the outlier value with their upper and lower limit instead of removing them.

###### Lets find out variables for outlier treatment

```python
for column in data.columns:
    if data[column].dtype == "int64" :
        plt.figure(figsize = (15,5))
        plt.title("Boxplot of "+ column)
        sns.boxplot(data = data,x = column)
        plt.show()
```

```python
attr_having_outliers = ['age', 'fnlwgt', 'hours-per-week']  #by seeing above boxplot 
```
